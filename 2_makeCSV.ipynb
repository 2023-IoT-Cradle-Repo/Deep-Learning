```python
import os
import pandas as pd

# 데이터 경로 지정
data_dirs = ["/content/drive/MyDrive/졸업작품/baby_data/bellypain",
             "/content/drive/MyDrive/졸업작품/baby_data/burping",
             "/content/drive/MyDrive/졸업작품/baby_data/discomfort",
             "/content/drive/MyDrive/졸업작품/baby_data/hungry",
             "/content/drive/MyDrive/졸업작품/baby_data/tired"]
# 데이터를 담을 리스트 생성
data = []

# data_dirs 리스트의 각 디렉토리에 대해 반복
for data_dir in data_dirs:
    # 해당 경로의 모든 wav 파일을 리스트에 저장
    wav_files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith(".wav")]

    # wav_files 리스트의 각 파일에 대해 반복
    for file in wav_files:
        # 파일 이름에서 label, id, count 정보 추출하여 리스트로 저장
        label, count = os.path.basename(file).replace(".wav", "").split("_")
        # 파일 이름에서 확장자 제거하여 파일 이름 저장
        file_name = os.path.basename(file)
        # 추출한 정보와 파일 이름을 데이터 리스트에 추가
        data.append([file_name, label])

# 데이터 리스트를 이용하여 DataFrame 생성
df = pd.DataFrame(data, columns=["file_name", "label"])
# DataFrame을 csv 파일로 저장
df.to_csv("/content/drive/MyDrive/졸업작품/datatrain.csv", index=False)
```

```python
import os
import pandas as pd

# 데이터 경로 지정
data_dirs = ["/content/drive/MyDrive/졸업작품/baby_data/bellypain",
             "/content/drive/MyDrive/졸업작품/baby_data/burping",
             "/content/drive/MyDrive/졸업작품/baby_data/discomfort",
             "/content/drive/MyDrive/졸업작품/baby_data/hungry",
             "/content/drive/MyDrive/졸업작품/baby_data/tired"]

# 데이터를 담을 리스트 생성
data = []

# label에 대응하는 숫자 딕셔너리 생성
label_to_number = {"bellypain": 0, "burping": 1, "discomfort": 2, "hungry": 3, "tired": 4}

# data_dirs 리스트의 각 디렉토리에 대해 반복
for data_dir in data_dirs:
    # 해당 경로의 모든 wav 파일을 리스트에 저장
    wav_files = [os.path.join(data_dir, file) for file in os.listdir(data_dir) if file.endswith(".wav")]

    # wav_files 리스트의 각 파일에 대해 반복
    for file in wav_files:
        # 파일 이름에서 label, id, count 정보 추출하여 리스트로 저장
        label, count = os.path.basename(file).replace(".wav", "").split("_")
        # 파일 이름에서 확장자 제거하여 파일 이름 저장
        file_name = os.path.basename(file)
        # 추출한 정보와 파일 이름을 데이터 리스트에 추가
        data.append([file_name, label, label_to_number[label]])

# 데이터 리스트를 이용하여 DataFrame 생성
df = pd.DataFrame(data, columns=["file_name", "label", "label_number"])

# DataFrame을 csv 파일로 저장
df.to_csv("/content/drive/MyDrive/졸업작품/datatrain2.csv", index=False)
```
