```python
import os

# 데이터셋 경로 설정
audio_folder = '/content/drive/MyDrive/졸업작품/baby_data'

# 클래스 폴더 리스트 생성
classes = os.listdir(audio_folder)

# 클래스 폴더별 wav 파일들의 이름 리스트 출력
for c in classes:
    files = os.listdir(os.path.join(audio_folder, c)) 
    print(f"{c} 폴더에 있는 파일 리스트: {files}")
    
    # 각 클래스 폴더 안의 wav 파일 이름을 변경하여 저장
    for i, f in enumerate(files):
        file_path = os.path.join(audio_folder, c, f)
        new_file_path = os.path.join(audio_folder, c, f"{c}_{i+1:03d}.wav")
        os.rename(file_path, new_file_path)

        
# {c}_{i+1:03d}
# 새로운 파일 이름은 클래스명_인덱스 형태로 저장됩니다. 
# 인덱스는 001부터 시작하며, 3자리 숫자로 맞춰집니다.
```
